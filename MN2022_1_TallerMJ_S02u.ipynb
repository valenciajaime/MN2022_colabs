```python
#importamos librerias
from scipy import optimize
import numpy as np
import math 
from matplotlib import pyplot as plt


#funcion del metodo de biseccion
def bisection(fun, a, b, error=None, iteraciones=10):
    """
    

    Parameters
    ----------
    fun : funcion a trabajar
    a : extremo inferior
    b :extremo superior
    error : difrerencia entre el resultado correcto y la solucion por el metodo
        
    iteraciones : numero de veces en la que se repite el proceso
        
    Returns
    retorna la raiz de la ecuacion, una grafica de la solucion y otra gráfica de las iteraciones vs el error

    """
    # el metodo de biseccion no puede ser aplicado
    if fun(a) * fun(b) >= 0:
        print(" el metodo de biseccion no puede ser aplicado")
        return None
    
    # calcular el numero de iteraciones sobre el error
    if error is not None:
        iteraciones = math.ceil(math.log((b - a) / error) / math.log(2))
    
    errores=[] #lista para almacenar los errores     
    num_iter=[] #lista para almacenar las iteraciones
    
    #así funciona el método
    for i in range(iteraciones + 1):
        x_m = (a + b) / 2
        
        if f(x_m) == 0:
            return x_m
        
        if f(a) * f(x_m) < 0:
            b = x_m
        else:
            a = x_m
    
        plt.plot(a,f(a),"o")
        plt.plot(b,f(b),"o")
       
        
        
        
        el_error= abs(x_m - optimize.bisect(fun, a, b))
        errores.append(el_error)
        num_iter.append(i)
 
    
    
    #Para las Graficas
    
    # Para la Gráfica 1
    x=np.linspace(-2,4,99999)
    y=f(x)
    plt.grid("on")
    g1=plt.plot(x, y)
    plt.title("Gráfica solucion")
    plt.xlabel("X")
    plt.ylabel("F(X)")
    plt.figure(1)
    plt.show
    
    #Para la Grafica 2
    plt.figure(2)
    plt.grid("on")
    g2=plt.plot(num_iter, errores)
    plt.title("Gráfica de iteracion vs error")
    plt.xlabel("ITERACIONES")
    plt.ylabel("ERROR")
    plt.show
   
  
    solucion = (a + b) / 2
 
    return print("la solucion encontrada por metodo de biseccion es", solucion )

f = lambda x: x**2 + 2*x - 8

print(bisection(f, 0, 5,1e-8))

print("el resultado utilizando la libreria scipy es: ", optimize.bisect(f, 0, 5))

```

```python
import math
from scipy import optimize
from matplotlib import pyplot
import numpy as np 



def f(x):
    return 5*x**3 - 5*x**2 + 6*x - 2


bis = optimize.bisect(f, 0, 1)

def bisection(fun, x_a, x_b, eps=None,steps=10):
    # el metodo no puede se aplicado 
    if fun(x_a) * fun(x_b) >= 0:
        print("no se puede aplicar este metodo")
        return None
    
    # calcular el numero de iteraciones segun el error o tolerancia 
    
    if eps is not None:
      
        steps = math.ceil(math.log((x_b - x_a) / eps) / math.log(2))
    
    XX=[] # lista de errores presentados 
    N=[] # lista de cantidad de iteraciones 
    
    #metodo de bisection
    for n in range(steps):
        x_m = (x_a + x_b) / 2
        
        if f(x_m) == 0:
            return x_m
        
        if f(x_a) * f(x_m) < 0:
            x_b = x_m
        else:
            x_a = x_m
            
        error = abs(x_m - optimize.bisect(fun, x_a, x_b))
        
        #grafica de puntos proximos a solución
        pyplot.plot(x_a,f(x_a),"o")
        pyplot.plot(x_b,f(x_b),"o")
        
        
        XX.append(error)
        
        N.append(n)
        
    
        
    x = np.linspace(-10,10,1000)
    y = f(x)
    pyplot.grid()
    pyplot.xlim(-15,15)
    pyplot.ylim(-15,15)
    pyplot.plot(x,y)
    pyplot.title("f(x)")
    pyplot.show()
    pyplot.grid()
    pyplot.plot(N,XX)
    pyplot.title("error")
    pyplot.show()
    
    
    
    return (x_a + x_b) / 2 , error

print(bisection(f, 0, 1, 1e-4))
```

```python
#-----------------------------------
# módulos utilizados
#-----------------------------------
import numpy as np#módulo para operaciones matriciales
from scipy import optimize as OP#módulo para llamar algorítmo de newton de scipy 
from scipy import misc #módulo para aplicar derivada
import matplotlib.pylab as plt#módulo para graficas
import math#módulo simular valor infinito
#-----------------------------------
# Método de Newton-Rapson
#-----------------------------------
def newton_rapson(f_root,p0,tol,N0,a,b):
    #-----------------------------------
    # Algorítmo de Newton-Rapson
    #-----------------------------------    
    i=0#contador iteraciones    
    error=tol+1# error inicial
    pnl=[]#lista vacia de apróximaciones
    el=[]#lista vacia de errores
    il=[]#lista vacia de iteraciones
    while i<N0 and error>tol:
        p = p0 - f_root(p0)/misc.derivative(f_root,p0)# Computar p_i
        error=abs(p-p0)#Calcular error como la diferencia entre aproximación actual y anterior
        i=i+1#aumento del contador de iteraciones 
        pnl.append(p0)#almaceno la aproximación anterior
        p0=p#igualo aproximación actual y anterior
        el.append(error)#almaceno error
        il.append(i)#almaceno valor del contador de iteraciones
    #-----------------------------------
    # Resultados del algotrimo propio
    #-----------------------------------
    print("Solución con algoritmo propio")
    print("n:",i)#imprimir número de iteraciones alcanzadas con el algoritmo propio
    print("pn:",p0)#imprimir raiz alcanzada
    print("Error:",error)#imprimir error alanzado de acuerdo a la tolerancia 
    #-----------------------------------
    # Resultados del algotrimo de scipy
    #-----------------------------------    
    print("Solución con el algoritmo del modulo scipy")
    Raiz = OP.newton(f_root,p0,rtol=tol)#utilizando función de scipy para el método de newton 
    print("pn:",Raiz)#imprimo el valor de la raiz obtenido con el modulo de scipy
    #-----------------------------------
    # grafica de puntos de iteracion sobre la función de trabajo
    #-----------------------------------    
    R_x = np.linspace(a,b)#lista de valores del eje x en la gráfica de la función de trabajo 
    Y_fr = f_root(R_x)    #lista de valores del eje y en la gráfica de la función de trabajo
    plt.plot(R_x,Y_fr,"-b")#gráfica de la función de trabajo
    plt.plot(pnl,f_root(pnl),"k.")#gráfica de aproximaciones a la raiz de la función de trabajo
    plt.xlim(a,b)#limites para el eje x
    plt.ylim(-10,10)#limites para el eje y
    plt.title("$f(x) = 0$")#titulo de la figura
    plt.xlabel("Eje x")#nombrando al eje x
    plt.ylabel("$y = f(x)$")#nombrando al eje y
    plt.grid("on")#habilotando rejilla
    plt.show()
    #-----------------------------------
    # grafica de error por cada iteración
    #-----------------------------------        
    plt.plot(il,el,"b-")
    plt.plot(il,el,"k.")
    plt.xlabel("n")#nombrando al eje x
    plt.ylabel("$Error_n$")#nombrando al eje y
    plt.grid("on")#habilotando rejilla
    plt.show()
    return None
#-----------------------------------
#Ejercicio 5.a. en sección 2.5 de Burde & Faires edición 9
#-----------------------------------
f_root = lambda x: np.power(x,3) - 2* np.power(x,2) - 5#función a trabajar propuesto en el ejercicio
a=1#límite x mínimo  propuesto en el ejercicio
b=4#límite x máximo  propuesto en el ejercicio
#-----------------------------------
p0=2#Aproximación inicial a la raiz 
N0=math.inf#Número máximo de iteraciones no definido en el ejercicio, debe alcanzar la toleracia. Se asume infinito
tol=1e-4#Tolerancia requerida en el ejercicio
#-----------------------------------
newton_rapson(f_root,p0,tol,N0,a,b)#llamando función creada con los parametros del ejercicio
```
