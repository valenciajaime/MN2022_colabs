##Usando odeint()

```python
#programacion funcion dy = f(t,y)
from scipy import integrate as IT

def Ejemplo2_BFc5(t,y):
  """ Ejemplo de Burden y Faires
      capitulo 5 ed9
      Ejemplo 2
  """
  #dy = -0.5*y
  dy = 1 + t*np.sin(t*y)
  return dy

## solucion usando odeint
t_init = 0
t_final = 2

y_0 = 3

T_sol = np.linspace(t_init,t_final,100) #100 instantes de calculo

Y_sol = IT.odeint(Ejemplo2_BFc5,y_0,T_sol, tfirst=True)
                  #por defecto dy/dt = f(y,t) se asume la primer variable
                                          # como la dependiente

##Grafico solucion

plt.plot(T_sol, Y_sol, "*-")
plt.plot(T_sol[0],y_0,"ro")
plt.xlabel("Eje t")
plt.ylabel("y(t)")
plt.grid("on")
plt.show()
```

##Usando solve_ivp()

```python
from scipy.integrate import solve_ivp


Y_sol1 = solve_ivp(Ejemplo2_BFc5,[t_init,t_final],[y_0])#,t_eval=T_sol)
                   #se asume la primera variable como la independiente por defecto
```

```python
#print(Y_sol1)

#grafico solucion usando solve_ivp
plt.plot(Y_sol1.t,Y_sol1.y[0],"b*-")
plt.plot(T_sol[0],y_0,"ro")
plt.xlabel("Eje t")
plt.ylabel("y(t)")
plt.grid("on")
plt.show()
```

## usando sympy.dsolve

```python
import sympy
from sympy import Function,dsolve,Derivative,Eq 
t=sympy.symbols('t') 
y=sympy.symbols('y', cls=Function) 
#ode=Eq(Derivative(y(t), t) + 9*y(t),0)
ode=Eq(Derivative(y(t), t) -1 - t*sympy.sin(sympy.Mul(t,y(t))),0) 
YS_sol = dsolve(ode, y(t),ics={y(0):3})
YS_sol
```
