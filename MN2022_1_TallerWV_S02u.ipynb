```python
#Numeral a) intervalo [0,1]:

import numpy as np
import matplotlib.pylab as plt
from math import *
from scipy import optimize as OP 

#Punto A)

print("Punto A):\n")

f = lambda x: (x**3)-(7*x**2)+(14*x)-6

def biseccion(f,a,b,tol):
  z = [a,b]
  w = z.copy()
  m1 = a
  m = b
  k = 0   #Contador de iteraciones

  l_error = []
  if f(a)*f(b)>0:
    print("La funci√≥n no cambia de signo")

  while abs(m1-m)>tol:
    m1 = m
    m = (a+b)/2
    if f(a)*f(m)<0: #Cambia de signo en el intervalo [a,m]
      b = m
    if f(m)*f(b)<0: #Cambia de signo en el intervalo [m,b]
      a = m
    print("El intervalo es [",a,",",b,"]")
    k = k + 1 
    error = abs(m1-m)

    plt.plot(m,f(m),"ro")

    l_error.append(error)

  R = OP.bisect(f,w[0],w[1])  #Usando Optimize

  print("\nN√∫mero de iteraciones:",k)

  #Punto B):

  print("\nPunto B):")

  print("\nSoluci√≥n usando Optimize:",R)
  print("Soluci√≥n con la funci√≥n creada:", m,"\n")
  print("El error de aproximaci√≥n entre la funci√≥n optimize y la creada es:",abs(m-R))

  #Punto C):

  print("\nPunto C):\n")

  R_x = np.linspace(w[0],w[1])

  Y_fr = f(R_x)

  plt.plot(R_x,Y_fr)

  plt.plot(m,f(m),"ro", label='Iteraciones')
  plt.plot(m,f(m),"k*", label='Soluci√≥n funci√≥n creada')
  plt.plot(R,f(R),"go", label='Soluci√≥n Optimize')

  plt.title("Gr√°fica iteraciones y raiz")
  plt.legend()  
  plt.xlabel("Eje x")
  plt.ylabel("y = f(x)")
  plt.grid("on")
  plt.show()


  #Punto D):

  print("\nPunto D):\n")  

  Rx1 = np.linspace(0, k, len(l_error))
  plt.figure(1)
  plt.title("Error por cada iteraci√≥n")
  plt.plot(Rx1, l_error, marker="o", label="Errores")

  plt.title("Gr√°fica error por cada iteraci√≥n")
  plt.legend()  
  plt.xlabel("Eje x")
  plt.ylabel("y = f(x)")
  plt.grid("on")
  plt.show()

  #Punto E)
  print("\nPunto E):\n")

  return m, error

m, error = biseccion(f,0,1,10**(-2))
print("Soluci√≥n obtenida:",m)
print("Error final:",error)

```

```python
import numpy as np
import matplotlib.pyplot as plt 

def Newton_Met(f, df, xizq , xder , toleracia):

  xini = xder

  iter = 0
  tabla = []
  puntos_plot = []
  puntos_error_plot = []
  y_plot=[]
  tramo = abs(2*tolerancia)

  while (tramo >= tolerancia):
    xnew = xini - (f(xini))/(df(xini))
    tramo = abs(xnew - xini)
    tabla.append([xini,xnew,tramo])
    y_plot.append(f(xini))
    puntos_plot.append(xini)
    puntos_error_plot.append(tramo)
    xini = xnew
    iter += 1

  #Pasamos la lista a un arreglo matricial
  tabla = np.array(tabla)
  n = len(tabla)

  print(["xini","xnew","tramo"])
  np.set_printoptions(precision = 4 )
  print(tabla)
  print("raiz en:", xini)
  print("Con margen de error de: ",tramo)

  x = np.linspace( (xizq +0.5 ), (xder) )
  plt.figure(1)
  plt.title("Raiz por metodo de Newton")
  plt.plot(x,f(x))
  plt.scatter(puntos_plot , y_plot)
  plt.grid()
  plt.show()

  iteraciones = np.linspace(0, iter, len(puntos_error_plot) )
  plt.figure(2)
  plt.title("Evolucion de los errores")
  plt.plot(iteraciones , puntos_error_plot)
  plt.scatter(iteraciones, puntos_error_plot)
  plt.grid()
  plt.show()


#Para el intervalo[0 , 1]
f =  lambda x : (np.exp(x)) - 3*(x**2)
df = lambda x :  (np.exp(x)) - 6*x
xder = 1
xizq = 0
tolerancia = 0.0001

Newton_Met(f, df, xizq , xder , tolerancia)
```

```python
#importaci√≥n de librerias 
import numpy as np
from matplotlib import pyplot as plt
from scipy.misc import derivative
from scipy import optimize as opt


def Soluci√≥n (f,tolerancia,Xi,Xj):
    

    # Punto A 
    
    # Intervalo de [-2 , 0]
    
    print("Punto A\n")    
    
    
    numero_iter = 0  # Contador de iteraciones 
    
    
    puntos_iter1= list()  # Creaci√≥n de listas con los valores de x por cada iteraci√≥n
    lista_iter1= list()   # Lista con la referencia de cada iteraci√≥n
    puntos_iter1.append(Xi)  
    

    # Ciclo que por cada iteraci√≥n se acerca a la ra√≠z

    while  f(Xi) < -tolerancia  or  f(Xi) > tolerancia:  
        
        
        Xi_mas1 = Xi - f(Xi)/(derivative(f,Xi,dx= 1e-6)) # M√©todo de Newton Raphson
        
        Xi = Xi_mas1 # Actualizaci√≥n de raiz
        
        numero_iter=  numero_iter+1  # Incremento de contador
        
        lista_iter1.append(numero_iter)  # Agrego cada iteraci√≥n a lista
        puntos_iter1.append(Xi) # Agrego cada valor de raiz a lista
    
    
    
    print(f"Raiz del problema: {Xi} \nCon un total de {numero_iter} iteraciones")
    
    
    # Para el intervalo de [0 , 2] repetimos el proceso
    
    
    Xj= 1
    
    numero_iter = 0
    
    puntos_iter2= list()
    lista_iter2= list()
    puntos_iter2.append(Xj)

    # Ciclo que por cada iteraci√≥n se acerca a la ra√≠z para el segundo intervalo
    
    while  f(Xj) < -tolerancia  or  f(Xj) > tolerancia:
        
        
        Xi_mas1 = Xj - f(Xj)/(derivative(f,Xj,dx= 1e-6))
        
        Xj = Xi_mas1
        
        numero_iter=  numero_iter+1
    
        puntos_iter2.append(Xj)
                
        lista_iter2.append(numero_iter)
    
    
    
    print(f"Raiz del problema: {Xj} \nCon un total de {numero_iter} iteraciones")
    
    
    
    
    # Punto B
    
    root1 = opt.bisect(f,-2,0)
    
    root2= opt.bisect(f,0,2)
    
    print("\nPunto B")
    print(f"Comprobaci√≥n del punto A con el m√©todo de bisecci√≥n:\n\nRaiz 1: {root1}\nRaiz 2: {root2}")
    
    
    # Punto C
    print("\nPunto C")
    print("Funci√≥n de iteraci√≥n sobre los puntos de trabajo\n")
    
    # print(puntos_iter1,puntos_iter2)
    

    X = np.linspace(-2,2)
    
    Y =  np.sin(X) + 3*np.cos(X) - 2
    
    Y1= 0*X
    
    # Gr√°fica con la soluci√≥n
    
    plt.plot(X,Y1,"--")
    plt.plot(puntos_iter1,f(puntos_iter1),"ro",label="Intervalo[-2,0]")
    plt.plot(puntos_iter2,f(puntos_iter2),"go",label="Intervalo[0,2]")
    plt.plot(X,Y,label="Funci√≥n",color="black")
    
    
    plt.title("M√©todo de Newton-Rapson")
    plt.xlabel("Eje x")
    plt.ylabel("Eje y")
    plt.legend(loc="lower center")
    plt.grid()
    plt.show()
    
    # Punto D 
    
    # Grafica de errores

    a= list(f(puntos_iter1[1:]))

    b= list(f(puntos_iter2[1:]))
  
    Lista1_errores= [abs(numero) for numero in a]    

    Lista2_errores= [abs(numero) for numero in b]    


    print("Punto D\n")
    plt.plot(lista_iter1,Lista1_errores,"-o", label = "Intervalo[-2,0]")
    plt.plot(lista_iter2,Lista2_errores,"-o",label = "Intervalo[0,2]")
    
    
    plt.title("Gr√°fica de errores")
    plt.xlabel("Error")
    plt.ylabel("Iteraci√≥n")
    plt.legend()
    plt.grid()
    plt.show()
    
    print("\n Punto E")
    return Xi,Xj,Lista1_errores[-1],Lista2_errores[-1]


# Valores ingresados en la funci√≥n

f = lambda x: np.sin(x) + 3*np.cos(x) - 2  # Funci√≥n a trabajar 
tolerancia= 5e-6
Xi= -2  # Valor elegido en el primer intervalo
Xj= 1



print(Soluci√≥n(f, tolerancia, Xi, Xj))
```
