# Primera aprox Numerica

```python
##Ejemplo1 pag. 174 Burden y Faires


F_ejem = lambda x:np.log(x)

dF_ejem = lambda x: 1/x

a = 1.8

df_a = dF_ejem(a)
print("Derivada en a: ", a, " = ",df_a)

h = 0.01

dnf_a = (F_ejem(a+h)-F_ejem(a))/h
print("Derivada Numerica en a: ", a, " = ",dnf_a, " con h= ",h)
```

# Curva de error

```python
# Limite numerico  por error de redondedo

F_ejem = lambda x:np.log(x) #funcion original
dF_ejem = lambda x: 1/x

h = 0.2
df_a = dF_ejem(a) #exacta

H = [h]
Df= [dnf_a]
for q in range(50):
  h = h/2
  dnf_a = (F_ejem(a+h)-F_ejem(a))/h
  H.append(h)
  Df.append(dnf_a)

Error = np.abs(np.array(Df)-df_a)

plt.figure(1)
plt.subplot(2,1,1)
plt.semilogx(H,Df,[H[-1],H[0]],[df_a,df_a],"ro-")
plt.xlabel("Eje h")
plt.ylabel("Deriva numerica")
plt.grid("on")
plt.legend(["Numerico","Derivada"])

plt.subplot(2,1,2)
plt.semilogx(H,Error)
plt.ylabel("Error")
plt.grid("on")

plt.show()
```

# funcion derivative() de scipy

```python
#Usando derivative de scipy

from scipy.misc import derivative

F_ejem = lambda x:np.log(x) #funcion original
dF_ejem = lambda x: 1/x

a = 1.8
h = 0.01
df_scipy = derivative(F_ejem,a, h)
df_a = dF_ejem(a)

print("Derivada exacta:", df_a)
print("Derivate(scipy): ",df_scipy)
```

```python
# Limite numerico  por error de redondedo
from scipy.misc import derivative

F_ejem = lambda x:np.log(x) #funcion original
dF_ejem = lambda x: 1/x

h = 0.1
a = 1.8
df_a = dF_ejem(a) #exacta

H = [h]
Df= [dnf_a]
for q in range(50):
  h = h/2
  dnf_a = derivative(F_ejem,a,h)
  H.append(h)
  Df.append(dnf_a)

Error = np.abs(np.array(Df)-df_a)

plt.figure(1)
plt.subplot(2,1,1)
plt.semilogx(H,Df,[H[-1],H[0]],[df_a,df_a],"ro-")
plt.xlabel("Eje h")
plt.ylabel("Deriva numerica")
plt.grid("on")
plt.legend(["Numerico","Derivada"])

plt.subplot(2,1,2)
plt.semilogx(H,Error)
plt.ylabel("Error")
plt.grid("on")

plt.show()
```

# funcion diff() de numpy

```python
##comando diff de numpy

F_ejem = lambda x:np.log(x) #funcion original
dF_ejem = lambda x: 1/x

x_dato = np.linspace(0.1,3,100)
y_f = F_ejem(x_dato)
y_df = dF_ejem(x_dato)

delta_x = np.diff(x_dato)#vectro n-1
delta_y = np.diff(y_f)
Dnum_f = delta_y/delta_x

plt.plot(x_dato,y_f)
plt.plot(x_dato,y_df)
plt.plot(x_dato[:-1],Dnum_f)
plt.grid("on")
plt.legend(["fun","Dfun","D_numerica"])
```
