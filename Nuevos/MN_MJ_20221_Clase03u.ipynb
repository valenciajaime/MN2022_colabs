Ecuacion No lineal de una variable


$1564000 = 1000000 e^{\lambda } +  \frac{435000}{\lambda}  \big( e^{\lambda} -1\big) $


**Funcion para busqueda de raiz:**

$f(\lambda) = -1564000 + 1000000 e^{\lambda } +  \frac{435000}{\lambda}  \big( e^{\lambda} -1\big) $

# **Grafico funcion**

```python

import numpy as np
import matplotlib.pylab as plt


# funcion para busqueda de raiz
f_root = lambda x: -1564000 + 1e6*np.exp(x) + (435000/x)*(np.exp(x)-1)

R_x = np.linspace(0.08,0.2)
Y_fr = f_root(R_x)

plt.plot(R_x,Y_fr)
plt.xlabel("Eje x")
plt.ylabel("y = f(x)")
plt.grid("on")
plt.show()
```

# Solucion de ecuacion No lineales (bisec-Newton)




[Module scipy.optimize](https://docs.scipy.org/doc/scipy/reference/optimize.html)

[Raices (roots)](https://docs.scipy.org/doc/scipy/reference/optimize.html#root-finding)($f(x) = 0$)

[Biseccion](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.bisect.html#scipy.optimize.bisect)

[Newton](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.newton.html#scipy.optimize.newton)

```python
# Solucion de la raiz

from scipy import optimize as OP #submodulo optimize

Raiz1 = OP.bisect(f_root,0.08,0.2)# metodo cerrado
print("Metodo Biseccion: ",Raiz1)

Raiz2 = OP.newton(f_root,0.2) #metodo abierto
print("Metodo de Newton: ",Raiz2)

np.abs(f_root(Raiz1))> np.abs(f_root(Raiz2))
```

```python
##Grafico solucion
plt.plot(R_x,Y_fr)
plt.plot(Raiz1,f_root(Raiz1),"k*")
plt.plot(Raiz2,f_root(Raiz2),"go")
plt.text(Raiz1-0.005,f_root(Raiz1),"Bisec", horizontalalignment='right')
plt.text(Raiz2+0.005,f_root(Raiz2),"Newton", horizontalalignment='left')
plt.title("$f(x) = 0$")
plt.xlabel("Eje x")
plt.ylabel("$y = f(x)$")
plt.grid("on")
plt.xlim(0.09,0.12)
plt.ylim(-10000,10000)
plt.show()
```

# Solucion de ecuacion No lineales (fixedpoint)

**Ecuacion No lineal de una variable**


$1564000 = 1000000 e^{\lambda } +  \frac{435000}{\lambda}  \big( e^{\lambda} -1\big) $


**Funcion para busqueda de raiz:**

$f(\lambda) = -1564000 + 1000000 e^{\lambda } +  \frac{435000}{\lambda}  \big( e^{\lambda} -1\big) $

**Funcion para Punto Fijo**

$f_{punto-fijo}(\lambda) = \frac{435000}{1000000} \frac{\big( e^{\lambda} -1\big)}{\big( 1.564 - e^{\lambda}\big)} $

[Punto Fijo (fixed point)](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fixed_point.html)($f_{fixed}(x) = x$)

```python
import numpy as np
import matplotlib.pylab as plt

# funcion para busqueda de raiz
f_root = lambda x: -1564000 + 1e6*np.exp(x) + (435000/x)*(np.exp(x)-1)

# funcion para busqueda de Punto Fijo
f_pf = lambda x:0.435*(np.exp(x)-1)/(1.564 - np.exp(x))

R_x = np.linspace(0.08,0.2)
Y_fpf = f_pf(R_x)

plt.plot(R_x,Y_fpf)
plt.title("$f_{fixed}(x) = x ?$")
plt.xlabel("Eje x")
plt.ylabel("y = f(x)")
plt.grid("on")
plt.show()
```

```python
# Solucion de la raiz

from scipy import optimize as OP

Pfijo1 = OP.fixed_point(f_pf,0.2) #metodo abierto
print("Metodo Punto Fijo: ",Pfijo1)

```

```python
##Grafico del punto fijo
plt.plot(R_x,Y_fpf)
plt.title("$f_{fixed}(x) = x$")
plt.plot(Pfijo1,f_pf(Pfijo1),"k*")
plt.xlabel("Eje x")
plt.ylabel(" f(x)")
plt.text(Pfijo1-0.005,f_pf(Pfijo1),"Fixed Point", horizontalalignment='right')
plt.grid("on")
plt.show()
```
