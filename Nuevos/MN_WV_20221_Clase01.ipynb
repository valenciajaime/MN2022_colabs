# Generar Datos

```python
x = np.linspace(0,4*np.pi,100)
y1 = np.cos(x)
y2 = np.cos(x+(np.pi/4))
y3 = np.sin(x)
```

# Graficar Datos

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```
