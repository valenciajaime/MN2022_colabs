# Metodo de Jacobi

```python
#Edicion de datos
A = [[10,-1,2,0],[-1,11,-1,3],[2,-1, 10, -1],[0, 3, -1, 8]] #Lista de listas

b = [[6],[25],[-11],[15]] #lista

A1 = np.array(A)
b1 = np.array(b)
print(A1)
print(b1)
```

```python
LL= -1*np.tril(A1,-1)
print("Triangula Inferior: \n",LL)

UU = -1*np.triu(A1,1)
print("Triangular Superior: \n",UU)

DD = A1+UU+LL
print("Diagonal: \n",DD)

M_Jacobi = np.dot(np.linalg.inv(DD), LL+UU)

print("Matriz Jacobi (T): \n",M_Jacobi)

c_Jacobi = np.dot(np.linalg.inv(DD),b1)
print("Vector ind Jacobi (c): \n",c_Jacobi)
```

```python
#Iteracion Jacobi paso-paso

X_inicio = np.array([[0], [0], [0],[ 0]]) #vector columna

X_new = np.dot(M_Jacobi,X_inicio) + c_Jacobi

print("Nuevo Valor: \n",X_new)
```

```python
#Esquema una iteracion de Jacobi
#Edicion de datos
A = [[10,-1,2,0],[-1,11,-1,3],[2,-1, 10, -1],[0, 3, -1, 8]] #Lista de listas

b = [[6],[25],[-11],[15]] #lista

A1 = np.array(A)
b1 = np.array(b)

X_inicio = np.array([[0], [0], [0],[ 0]]) #vector columna

X_new = np.zeros(X_inicio.shape)
N = len(b1)
for i in range(N):
  Tsum = 0
  for j in range(N):
    if i!=j:
      Tsum = Tsum - A1[i,j]*X_inicio[j]
  Tsum = (Tsum + b1[i])/A1[i,i]
  X_new[i] = Tsum

print(X_new)
```

# Funcion Iteracion Jacobi

```python
#Edicion de datos
import numpy as np
import matplotlib.pylab as plt

A = [[10,-1,2,0],[-1,11,-1,3],[2,-1, 10, -1],[0, 3, -1, 8]] #Lista de listas

b = [[6],[25],[-11],[15]] #lista

A1 = np.array(A)
b1 = np.array(b)
print(A1)
print(b1)

X_inicio = np.array([[0], [0], [0],[ 0]]) #vector columna


def Iter_Jacobi(AA, bb, X0):
  """Funcion de una iteracion de Jacobi
     sistema AA x = bb
     AA: matriz (np.array(n,n))
     bb: vector (np.array(n,1))
     X0: vector )np.array(n,1))
  """
  X_new = np.zeros(X0.shape)
  N = len(bb)
  for i in range(N):
    Tsum = 0
    for j in range(N):
      if i!=j:
        Tsum = Tsum - AA[i,j]*X0[j]
    Tsum = (Tsum + bb[i])/AA[i,i]
    X_new[i] = Tsum
  return X_new


X1 = Iter_Jacobi(A1,b1,X_inicio)
print("X1: \n",X1)

X2 =  Iter_Jacobi(A1,b1,X1)
print("X2: \n",X2)

X3 =  Iter_Jacobi(A1,b1,X2)
print("X3: \n",X3)
```

# Funcion Metodo Jacobi

```python
###Esquema Iteracion 
def Metodo_Jacobi(AA,bb,X0, Max_Iter=10, Tol=0.0001):
  """Metodo de Jacobi para solucion iterativa de
     AA x = bb
     
  """
  X_old = X0
  C_iter = 1 #contador iteraciones
  C_error = 1 #error inicial
  while C_iter<Max_Iter and C_error>Tol:
    X_new = Iter_Jacobi(AA,bb, X_old)
    print("Iteracion: ",C_iter)
    print("Vector: ",X_new)

    ##incremento contador
    C_iter += 1
    #Nuevo Error
    C_error = np.linalg.norm(X_new - X_old)
    print("Error: ",C_error)

    #reiniciar
    X_old = X_new
  print("Sale ciclo while")
  return X_new, C_iter, C_error

Metodo_Jacobi(A1,b1,X_inicio, Max_Iter=10, Tol=0.5)
```
