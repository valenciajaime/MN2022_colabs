# Calculo numerico

```python
##Ejemplo1 pag. 198 Burden y Faires
#Calculo de la integral entre 2 y 2

import scipy.integrate as IT

F_a = lambda x: x**2
F_b = lambda x: x**4
F_c = lambda x: 1/(x+1)


#teniendo la funcion explicita
I_a = IT.quad(F_a,0,2)
print("integral Fa (quad): ",I_a)

I_b = IT.quad(F_b,0,2)
print("integral Fb (quad): ",I_b)

I_c = IT.quad(F_c,0,2)
print("integral Fc (quad): ",I_c)
```

```python
#teniendo solo datos discretos de la funcion

x_dato = np.linspace(0,2,10)

y_fa = F_a(x_dato)
y_fb = F_b(x_dato)
y_fc = F_c(x_dato)

In_a = IT.trapz(y_fa,x_dato)
print("integral Fa (trapz): ",In_a)#*(2/9))

In_b = IT.trapz(y_fb,x_dato)
print("integral Fb (trapz): ",In_b)

In_c = IT.trapz(y_fc,x_dato)
print("integral Fc (trapz): ",In_c)
```

```python
#teniendo solo datos

x_dato = np.linspace(0,2,10)

y_fa = F_a(x_dato)
y_fb = F_b(x_dato)
y_fc = F_c(x_dato)

In_as = IT.simps(y_fa,x_dato)
print("integral Fa (simps): ",In_as)

In_bs = IT.simps(y_fb,x_dato)
print("integral Fb (simps): ",In_bs)

In_cs = IT.simps(y_fc,x_dato)
print("integral Fc (simps): ",In_cs)
```

# Calculo simbolico

```python
# 

x = SP.Symbol("x")

Fs_a = x**2
Fs_b = x**4
Fs_c = 1/(x+1)

IFs_a=Fs_a.integrate()
print("Antiderivada a: ",IFs_a)

IFs_b=Fs_b.integrate()
print("Antiderivada b: ",IFs_b)

IFs_c=Fs_c.integrate()
print("Antiderivada c: ",IFs_c)

```

```python
IFn_a = SP.lambdify(x, IFs_a) #Creacion de funcion evaluable
Ia_exacta = IFn_a(2)-IFn_a(0)
print("integral Fa (exacta): ",Ia_exacta)


IFn_b = SP.lambdify(x, IFs_b) #Creacion de funcion evaluable
Ib_exacta = IFn_b(2)-IFn_b(0)
print("integral Fb (exacta): ",Ib_exacta)

IFn_c = SP.lambdify(x, IFs_c) #Creacion de funcion evaluable
Ic_exacta = IFn_c(2)-IFn_c(0)
print("integral Fc (exacta): ",Ic_exacta)
```
