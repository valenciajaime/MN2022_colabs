# Desfaces

[Animacion desfaces en elementos R, L, C: ](https://www.animations.physics.unsw.edu.au/jw/AC.html#animation6)

```python
#oscilogramas

t = np.arange(0,0.1,1e-3)

f_hz = 60 #Frecuencia(numero de veces que se repite la onda en un seg)

wo = 2*np.pi*f_hz #Frecuencia angular

V_t = 155*np.cos(wo*t) #Onda de voltaje

A_desf = -wo*(3e-3) #Angulo negativo retrazo de la onda
I_t = 44*np.cos(wo*t + A_desf) #onda de corriente desfasada

plt.figure(1)
plt.plot(t,V_t, t, I_t)
plt.legend(["Voltaje","Corriente"])
plt.grid("on")
plt.xlabel("Tiempo(s)")

plt.figure(2)
plt.plot(V_t,I_t)
plt.xlabel("V_t")
plt.ylabel("I_t")

plt.show()

```

# Sistema lineal complejo

Fasores_2022-05-26 15-12-06.png

[Circuitos AC: ](https://www.philadelphia.edu.jo/academics/fobeidat/uploads/Electric%20Circuits%20II%20Course/5%20Phasor%20diagram.pdf)

```python
#sistema lineal complejo

E =complex(100,0) #Fuente de voltaje
Z1 = complex(4,0) # Rama Resistiva
Z2 = complex(9,-7) # Rama capacitiva
Z3 = complex(8,6) # Rama inductiva

A = np.array([[Z1,Z2],[Z3, -(Z2+Z3)]], dtype = np.complex_)
b = np.array([E,0], dtype = np.complex_)

I = np.linalg.solve(A,b)

print("I: ",I[0])
print("Magnitud I: ",abs(I[0]))
print("Fase I: ",np.angle(I[0],deg=True))
```
