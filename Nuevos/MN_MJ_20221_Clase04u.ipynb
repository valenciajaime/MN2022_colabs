# **Grafico max y min**

```python

import numpy as np
import matplotlib.pylab as plt


# funcion
## polinomio: x**4 - 21 x**3 + 122 x**2 - 64 x  - 208

P1 = np.array([1, -21, 122, -64, -208 ])

Raiz = np.roots(P1)
print("Raices: ",Raiz)
x = np.linspace(-2,11,200)
y_p = np.polyval(P1,x)

plt.plot(x,y_p)
plt.grid("on")
plt.xlabel("Eje x")
plt.ylabel("f(x)")
plt.show()
##como obtener los maximo o minimos locales??
```

```python
##Calculo de puntos criticos
dP1 = np.polyder(P1)
P_criticos = np.roots(dP1)
y_pcriticos = np.polyval(P1,P_criticos)

plt.figure(1)
plt.plot(x,y_p)
plt.plot(P_criticos, y_pcriticos,"ro")
#plt.show()

## Maximos o minimos??
ddP1 = np.polyder(P1,2) #segunda derivada
y_ddcriticos = np.polyval(ddP1, P_criticos)
p_min = y_ddcriticos>0

plt.figure(2)
plt.plot(x,y_p)
plt.plot(P_criticos, y_pcriticos,"ro")
plt.grid("on")
plt.xlabel("Eje x")

for q in range(3):
    if p_min[q]==True:
        plt.text(P_criticos[q],y_pcriticos[q],"Minimo")
    else:
        plt.text(P_criticos[q],y_pcriticos[q],"Maximo")
```

# Max == -Min

```python
import numpy as np
import matplotlib.pylab as plt


# funcion
## polinomio: x**4 - 21 x**3 + 122 x**2 - 64 x  - 208

P1 = -1*np.array([1, -21, 122, -64, -208 ]) #SE MULTIPLICA POR -1

x = np.linspace(-2,11,200)
y_p = np.polyval(P1,x)

##Calculo de puntos criticos
dP1 = np.polyder(P1)
P_criticos = np.roots(dP1)
y_pcriticos = np.polyval(P1,P_criticos)

plt.figure(1)
plt.plot(x,y_p)
plt.plot(P_criticos, y_pcriticos,"ro")
#plt.show()

## Maximos o minimos??
ddP1 = np.polyder(P1,2) #segunda derivada
y_ddcriticos = np.polyval(ddP1, P_criticos)
p_min = y_ddcriticos>0

plt.figure(2)
plt.plot(x,y_p)
plt.plot(P_criticos, y_pcriticos,"ro")
plt.grid("on")
plt.xlabel("Eje x")

for q in range(3):
    if p_min[q]==True:
        plt.text(P_criticos[q],y_pcriticos[q],"Minimo")
    else:
        plt.text(P_criticos[q],y_pcriticos[q],"Maximo")

```
